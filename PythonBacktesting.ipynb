```python
import numpy as np
from yahoo_fin.stock_info import get_data
import yahoo_fin.stock_info as si
from datetime import datetime, timedelta
import plotly.express as px
import pandas as pd
```

```python
nasdaq_list = {"AAPL", "AMZN", "MSFT"}
historic_data = {}
for i in nasdaq_list:
    print(i)
```

```python
for ticker in nasdaq_list:
    historic_data[ticker] = si.get_data(ticker, start_date="01/01/2013", end_date="01/03/2023", interval="1mo")

historic_data
```

```python
nasdaq_stats ={}
for ticker in nasdaq_list:
    temp = si.get_stats_valuation(ticker)
    temp = temp.iloc[:,:2]
    temp.columns = ["Attribute", "Recent"]
    nasdaq_stats[ticker] = temp

nasdaq_stats
```

```python
combined_stats = pd.concat(nasdaq_stats)
combined_stats = combined_stats.reset_index()
combined_stats
```

```python
market_caps = combined_stats[combined_stats["Attribute"]=="Market Cap (intraday)"]
del market_caps["level_1"]
market_caps.columns = "Ticker", "Attribute", "Value"
market_caps
```
